```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

## Goal

You work for a data science consulting company. A major video game production company, has retained your firm to conduct market research into the video game industry. They've furnished you with video game sales data for the last thirty years (described below) and, as a first project, would like to know:

1. What are some major differences between the North American, European, and Japanese video game markets?

2. What video game genres are trending in each market? 

3. What features about a video game are most indicative of its success?

```python
vgsales = pd.read_csv("./data/vgsales.csv")
```

```python
vgsales.head()
```

```python
vgsales.shape
```

```python
vgsales.count()
# so missing a little bit of data, not a lot
# Just drop for now?
```

```python
categorical_variables = ['Platform','Genre']
```

```python
for column in categorical_variables:
    sns.countplot(x=column, data=vgsales)
    plt.xticks(rotation=90)
    plt.show()
```

```python
#vgsales.Publisher.value_counts()
```

```python
vgsales['NA_Sales_pct'] = vgsales.NA_Sales / vgsales.Global_Sales
vgsales['EU_Sales_pct'] = vgsales.EU_Sales / vgsales.Global_Sales
vgsales['JP_Sales_pct'] = vgsales.JP_Sales / vgsales.Global_Sales
```

```python
bp = vgsales[['Genre','NA_Sales_pct']].boxplot(by='Genre')
bp.get_figure().gca().set_xlabel("")
bp.get_figure().gca().set_title("")
plt.xticks(rotation=90)
bp
```

```python
bp = vgsales[['Genre','EU_Sales_pct']].boxplot(by='Genre')
bp.get_figure().gca().set_xlabel("")
bp.get_figure().gca().set_title("")
plt.xticks(rotation=90)
bp
```

```python
bp = vgsales[['Genre','JP_Sales_pct']].boxplot(by='Genre')
bp.get_figure().gca().set_xlabel("")
bp.get_figure().gca().set_title("")
plt.xticks(rotation=90)
bp
```
