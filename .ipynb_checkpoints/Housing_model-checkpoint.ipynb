```python
import warnings
import pickle

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from pprint import pprint
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, cross_val_score

%matplotlib inline

warnings.filterwarnings('ignore')
```

```python
# downloaded here https://www.kaggle.com/harlfoxem/housesalesprediction
kc_data = pd.read_csv("/Users/codyschank/data_challenges/data/kc_house_data.csv")
```

```python
# tick marks at 1 million, 2 million...
# if we were using regression, we would want to consider the distribution of these values. 
# At very least, a Tobit regression. But maybe also something to account for extreme outliers.
plt.figure(figsize=(8,4))
sns.distplot(kc_data.price)
plt.show()
```

```python
#kc_data.head()
#kc_data.dtypes
kc_data.price.mean()
```

```python
kc_data.price.median()
```

```python
# convert date to month, could also do week...but would take a little extra work
kc_data['month'] = pd.to_numeric(kc_data.date.str[4:6])
```

```python
# drop just the id and date field
kc_data_select = kc_data.drop(columns=['id','date'])
kc_data_select.shape
# kc_data_select.head()
```

```python
X = kc_data_select.drop(columns=['price'])
y = kc_data_select.price
```

```python
# plot the features
plt.figure()
plt.clf()
X.hist(figsize=(12,6))

plt.tight_layout()
```

### Fit on all of the data

```python
rf = RandomForestRegressor(random_state=42, n_estimators=800, max_features="log2")
rf.fit(X, y)
```

```python
feature_importances = pd.DataFrame(X.columns.values, columns=['feature'])
feature_importances['importance'] = [round(i,3) for i in rf.feature_importances_]
print(feature_importances.sort_values(by=['importance'], ascending=False))
```

```python
gb = GradientBoostingRegressor(max_depth=5, n_estimators=10000, n_iter_no_change=500)
gb.fit(X, y)
```

```python
feature_importances = pd.DataFrame(X.columns.values, columns=['feature'])
feature_importances['importance'] = [round(i,3) for i in gb.feature_importances_]
print(feature_importances.sort_values(by=['importance'], ascending=False))
```

### Pickle

```python
# save the model to disk
# why is it such a big file?
filename = '/Users/codyschank/data_challenges/models/finalized_model.sav'
pickle.dump(rf, open(filename, 'wb'))
```

```python
# save dataframe means to disk
filename = '/Users/codyschank/data_challenges/models/data_means.sav'
pickle.dump(X.mean(axis = 0), open(filename, 'wb'))
```

### Crossvalidation

```python
rf_cv_r2 = cross_val_score(rf, X, y, cv=5, scoring = 'r2')
rf_cv_mae = cross_val_score(rf, X, y, cv=5, scoring = 'neg_median_absolute_error')
```

```python
print(np.mean(gb_cv_mae))
```

### GridSearchCV to find best hyperparamters in Random Forest Regressor

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 5)]
# Number of features to consider at every split
max_features = ['sqrt', 'log2']

# Create the random grid
param_grid = {'n_estimators': n_estimators, 'max_features': max_features}

pprint(param_grid)
```

```python
rf = RandomForestRegressor()
# Instantiate the grid search model
grid_search_rf = GridSearchCV(estimator = rf, param_grid = param_grid,  cv = 5, n_jobs = 6, verbose = 2, scoring = 'r2')
```

```python
# Fit the grid search to the data
grid_search_rf.fit(X, y)
```

```python
grid_search_rf.best_params_
```

```python
grid_search_rf.best_score_
```

### GridSearchCV to find best hyperparamters in Gradient Boosting Regressor

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10000, stop = 70000, num = 7)]
# Number of features to consider at every split
max_depth = [1,2,3,4,5,6,7]

# I went a bit overboard here, reduce max depth? institute stopping rule?

# Create the random grid
param_grid = {'n_estimators': n_estimators, 'max_depth': max_depth}

#pprint(param_grid)
```

```python
gb = GradientBoostingRegressor(n_iter_no_change=500)
# Instantiate the grid search model
grid_search_gb = GridSearchCV(estimator = gb, param_grid = param_grid,  cv = 5, n_jobs = 6, verbose = 2, scoring = 'r2')
```

```python
# Fit the grid search to the data
grid_search_gb.fit(X, y)
```

```python
grid_search_gb.best_params_
```

```python
# interesting that max_depth and n_estimators are both the highest here. I guess I should try some more settings? Let's set the stopping rule. 
```
